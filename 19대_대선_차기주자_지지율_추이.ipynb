Following his retirement as UN Secretary-General, Ban Ki-moon emerged as a strong presidential contender for the Saenuri Party and was seen as a serious rival to Moon Jae-in. However, around October 2016, the public backlash triggered by the Choi Soon-sil scandal and the growing movement for the impeachment of President Park Geun-hye significantly weakened the momentum of the conservative party. Subsequently, Ban Ki-moon abruptly announced his decision not to run on February 1, 2017.

After his withdrawal, Acting President Hwang Kyo-ahn briefly emerged as the new frontrunner for the Saenuri Party but also declared that he would not run in March. As a result, Hong Joon-pyo took over as the party’s presidential candidate.

Within the Democratic Party, Moon Jae-in continued to lead as the dominant candidate. Nevertheless, rising political figures such as Lee Jae-myung and Ahn Hee-jung showed consistent performances, fearlessly following behind Moon. At the time, Lee was serving as the Mayor of Seongnam, and Ahn was the Governor of South Chungcheong Province. Notably, before the party primary concluded, Ahn Hee-jung at one point surpassed both Lee Jae-myung and Ahn Cheol-soo in support within the broader opposition bloc.

It becomes clear that gaining substantial public exposure is a prerequisite for rising as a major presidential contender. For example, Hong Joon-pyo was serving as the Governor of South Gyeongsang Province at the time; Ban Ki-moon, who appeared as a “relief pitcher” candidate, had previously served as Secretary-General of the United Nations; and Hwang Kyo-ahn also emerged as a leading conservative figure due to his roles as Prime Minister and Acting President.

### consideration

However, it is important to note that the 19th presidential election was an early election triggered by the impeachment of President Park Geun-hye. As such, with the exception of President-elect Moon Jae-in, prior experience in relevant government positions appeared to play a more significant role than media exposure in shaping public support for other candidates.

1. The Consistently Flat Support for Yoo Seung-min Since May 2016
Yoo Seung-min maintained a steady level of public support starting around May 2016, but his approval ratings neither increased nor decreased significantly over time. This plateau can be attributed to his limited public exposure and lack of a dramatic political turning point that could boost his profile. Although he was recognized for his policy expertise and principled stance, these traits alone were not sufficient to expand his support base in a volatile and media-driven political climate.

2. Media Exposure of President-elect Moon Jae-in
During the pre-election period, Moon Jae-in enjoyed a high degree of media exposure, which played a crucial role in reinforcing his public image and political momentum. As a former Chief Presidential Secretary under President Roh Moo-hyun and the previous presidential runner-up in 2012, Moon had already built substantial name recognition. His continuous presence in the media, particularly in connection with the impeachment of President Park Geun-hye and the calls for political reform, positioned him as a stable and experienced leader in the eyes of the public.

3. Patterns in Past Presidential Elections
A review of previous presidential elections in South Korea reveals several trends among successful candidates. Most notably, with the exception of Lee Myung-bak, none of the recent presidents—including Kim Young-sam, Kim Dae-jung, Roh Moo-hyun, Park Geun-hye, Moon Jae-in, and Yoon Suk-yeol—had prior experience as a local government head.

Kim Young-sam and Kim Dae-jung were long-time political activists and prominent figures in the democratization movement, which gave them widespread recognition well before their presidential bids. Roh Moo-hyun, although relatively less known at the national level at first, gained rapid public support through his reformist image and effective media engagement.

Lee Myung-bak’s case was different: after serving two terms as a National Assembly member, he became Mayor of Seoul and received high praise for his administrative performance, which directly contributed to his successful presidential campaign.

Park Geun-hye, on the other hand, served five terms in the National Assembly and benefited from strong name recognition as the daughter of former President Park Chung-hee—a legacy that played a significant role in her political ascent.
