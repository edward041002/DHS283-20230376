Yoon Suk-yeol first rose to national prominence when President Moon Jae-in appointed him to lead the investigation into the “Choi Soon-sil scandal,” which involved former President Park Geun-hye. He later prosecuted high-profile figures such as Samsung’s Lee Jae-yong and former President Lee Myung-bak. Yoon further solidified his public image as a prominent prosecutor by taking on other major cases, including the “Burning Sun scandal.” In July 2019, he was appointed as Prosecutor General, again by President Moon.

However, tensions grew when Cho Kuk, a close aide to President Moon, was nominated as Minister of Justice. Around the time Cho began to receive attention as a potential presidential candidate, allegations of corruption involving his daughter surfaced, and Yoon initiated a formal investigation. This marked a turning point where Yoon, though appointed by the current administration, began investigating its inner circle.

Yoon never explicitly expressed political ambitions during his term, but he frequently clashed with the Moon administration, gaining increasing attention as a potential political figure.

On March 4, 2021, Yoon resigned from his post as Prosecutor General. Later, on June 29, he officially declared his candidacy for the presidency as an independent. A month later, on July 30, he joined the conservative People Power Party (PPP).

Following his resignation, public anticipation of his presidential bid grew, leading to a sharp rise in his approval ratings. He eventually became one of the two leading candidates, competing directly with Democratic Party candidate Lee Jae-myung.

It is worth noting that, in the early stages, Hwang Kyo-ahn and Lee Nak-yon initially led the polls as the most prominent presidential hopefuls within the conservative and liberal blocs, respectively. However, both saw a noticeable decline in momentum at specific turning points.

Lee Nak-yon's early approval ratings showed a steady upward trend as the leading candidate from the ruling bloc. However, his momentum weakened with the emergence of Cho Kuk as a political figure, and while Lee appeared to regain strength in response to the corruption scandals surrounding Cho, his rise was eventually overshadowed by the growing popularity of Lee Jae-myung. This culminated in Lee Jae-myung defeating him in the Democratic Party's primary.

In the case of former Prime Minister Hwang Kyo-ahn, it is particularly notable that he initially garnered higher approval ratings than Hong Joon-pyo, the Saenuri Party’s presidential candidate who had run against Moon Jae-in in the 19th election. This suggests there was considerable anticipation among conservative voters regarding Hwang’s potential as a presidential contender. However, as Yoon Suk-yeol began to gain national attention as a strong opponent to the Moon administration, Hwang’s support base quickly eroded. Remarkably, Yoon overtook Hwang in conservative approval ratings even before formally announcing his candidacy.

During this period, Lee Nak-yon held several high-profile positions. He served as Prime Minister under President Moon Jae-in starting in May 2017, was elected as a National Assembly member representing Jongno District in the 21st general election in 2020, and later became the leader of the Democratic Party from August 2020 to March 2021.

Notably, Lee’s approval ratings began to decline from the time he assumed the party leadership—despite having held strong support in the early part of 2020. His downward trend became more pronounced following a New Year’s interview in January 2021, in which he expressed support for the pardon of former presidents Park Geun-hye and Lee Myung-bak. While seemingly aimed at appealing to centrist voters, the statement backfired politically, and Lee Jae-myung overtook him in popularity soon afterward.

In the case of Lee Jae-myung, his political future faced serious challenges due to multiple legal controversies. These included allegations of abuse of power related to the forced hospitalization of his older brother in 2019, charges of spreading false information under the Public Official Election Act regarding the Daejang-dong development scandal, and similar charges concerning a past case of impersonating a prosecutor. As a result, despite having been elected Governor of Gyeonggi Province in 2018, he came close to losing his position, and his public support—once placing him as a notable contender behind Moon Jae-in and Ahn Hee-jung in the previous election—significantly declined.

However, on July 16, 2020, the Grand Bench of the Supreme Court overturned his conviction regarding the dissemination of false information, effectively allowing him to retain his governorship. Entering 2021, Lee's popularity surged once again, eventually surpassing that of Lee Nak-yon. In the Democratic Party's primary race, he secured 50.29% of the votes, officially becoming the party’s presidential nominee.



### An Examination of the Key Attributes of Presidential Candidates

It appears that public dissatisfaction with the current administration often translates into support for figures who are perceived as having most fiercely opposed it. Of course, mere opposition is not enough to draw meaningful attention. For instance, former Prime Minister Hwang Kyo-ahn had previously announced that he would not run for president during his tenure as Acting President, in the interest of national stability—a move that earned him a measure of respect. Similarly, former Prosecutor General Yoon Suk-yeol gained public support by building an image of principled persistence, continuing his investigations even though he had been appointed by the president himself.

In summary, figures who emerge as major opposition presidential contenders tend to meet two key criteria:

They have engaged in visible and intense confrontation with the current administration, or inspire public curiosity about what might have happened had they run in the previous election.

They either possess substantial administrative experience or display qualities that cast them in a heroic or resolute light.

Within the ruling party, presidential hopefuls often emerge from two main backgrounds: those who drew significant attention in the previous presidential election, or those who have held key positions within the current administration.

Lee Jae-myung, for instance, was a notable figure in the previous election alongside Ahn Hee-jung. After Ahn's downfall due to the MeToo movement, Lee garnered even greater attention as a leading candidate. His appearances on television programs such as Same Bed, Different Dreams helped him build wider public recognition and a more relatable, positive image.

Lee Nak-yon, on the other hand, served in several high-level roles within the Moon administration, including as Prime Minister and later as leader of the Democratic Party. His steady rise as a presidential contender can likely be attributed to the public visibility and authority that came with these positions.

Reflecting on the 19th presidential election as well, it appears that the qualities defining a viable presidential contender lie less in political connections among elites, and more in factors such as public exposure, prior experience in administrative or high-ranking positions, and the ability to convey a consistent political vision or conviction.

From this perspective, the traits that appeal to voters differ by political alignment. For supporters of the ruling party, a central question seems to be whether a candidate has been a key figure within the current administration—such as Lee Nak-yon. In contrast, opposition supporters tend to gravitate toward candidates who either left a strong impression in the previous election without winning (e.g., Hwang Kyo-ahn, Moon Jae-in), or those who represent a powerful antithesis to the current administration—such as Yoon Suk-yeol.
